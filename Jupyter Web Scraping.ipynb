# 3 Write functions to grab required text 

```python
def make_soup(single_url):
    # Given a URL, make soup
    base_url = requests.get(single_url)
    object_ = base_url.text
    soup_ = BeautifulSoup(object_, features='lxml')#'lxml') #storing into a variable
    return soup_
```

```python
# Testing to see text
soup=make_soup(list_of_urls[0])
text_=soup.text
print(text_)
```

```python
# Testing to see contents
contents_=soup.contents
print(contents_)
```

```python
# With the recursifor child in soup.recursiveChildGenerator().
# This goes through the document tree and prints the names of all HTML tags
for child in soup.recursiveChildGenerator():
    if child.name:
        print(child.name)
```

```python
# Create a set of unique tags
unique_tags=[]
for child in soup.recursiveChildGenerator():
    if child.name:
        unique_tags.append(child.name)
unique_tags=set(unique_tags)
unique_tags=list(unique_tags)
unique_tags.sort()
print(unique_tags)
```

# 4 Web Scraping

```python
print('Soup Title:               ', soup.title)
print('Soup Title Text:          ', soup.title.text)
print('Soup Title Text Parent:   ', soup.title.parent)
```

```python
# Find and print all tags by category

for tag_type in unique_tags:
    print("")
    print("Printing all tags, for tag type", tag_type)
    print("")
    for tag in soup.find_all(tag_type):
        print("{0}: {1}".format(tag.name, tag.text))
```

```python
# Finds all h2 and p elements and prints their text.

tags = soup.find_all(['h2', 'p'])
for tag in tags:
    print(" ".join(tag.text.split())) 
```

```python
# Finds all h3 and p elements and prints their text.

tags = soup.find_all(['h3', 'p'])
for tag in tags:
    print(" ".join(tag.text.split())) 
```

```python
# Finds all h4 and p elements and prints their text.

tags = soup.find_all(['h4', 'p'])
for tag in tags:
    print(" ".join(tag.text.split())) 
```
